```python
# Name reversing and splitting if uppercase
text = input("Enter a string: ")

if len(text) == 0:
    print("Empty string.")
else:
    mid = len(text) // 2
    first_half = text[:mid]
    second_half = text[mid:]

    if text.isupper():
        first_half = first_half[::-1]
        second_half = second_half[::-1]
        print(f"First half reversed: {first_half}")
        print(f"Second half reversed: {second_half}")
    else:
        print("String is not fully uppercase. No reversal applied.")
        print(f"First half: {first_half}")
        print(f"Second half: {second_half}")
```

```python
# Write a program that:

# Takes an encoded message as input
# Decodes it using the steps above in reverse:

# Convert to lowercase
# Extract every 2nd character
# Reverse the result


# Prints the decoded message

# Example:
# Input:  RDAEES
# Output: seed

encoded = input("Enter encoded message: ")
step1 = encoded.lower()
print("Step 1 - lowercase:", step1)
step2 = step1[::2]
print("Step 2 - every 2nd char:", step2)
step3 = step2[::-1]
print("Step 3 - reversed:", step3)
print("Decoded message:", step3)
```

```python
# Problem: Initials Extractor
# Write a program that:

# Takes a full name as input (e.g. Hanan Ahmad Rawa)
# Extracts and prints the initials like this:

# ============================
#      INITIALS EXTRACTOR
# ============================
# Full Name : Hanan Ahmad Rawa
# Initials  : H.A.R
# ============================
# Rules:

# Use string slicing and indexing only
# No loops, no split() — just what you know
# Assume the name always has exactly 3 words
full_name = input("Enter your full name: ")
first = full_name[0]
space1 = full_name.index(" ")
middle = full_name[space1 + 1]
space2 = full_name.index(" ", space1 + 1)
last = full_name[space2 + 1]
initials = first + "." + middle + "." + last
print("============================")
print("INITIALS EXTRACTOR".center(28))
print("============================")
print("Full Name : " + full_name)
print("Initials  : " + initials)
print("============================")
```
