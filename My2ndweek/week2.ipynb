```python
# Write a program that:

# Takes a password as input
# Checks and prints the following about it:

# ============================
#    PASSWORD ANALYSIS REPORT
# ============================
# Password        : hanan123
# Length          : 8
# Has digits      : True
# Is alphanumeric : True
# Is all lower    : True
# Is all upper    : False
# ============================
Pw = input("Enter your password:")
print("============================")
n = "PASSWORD ANALYSIS REPORT"
print(n.center(28))
print("============================")
print("Password", Pw.center(22))
l = int(len(Pw))
print("Length:",l)
print("Has digits:",any(char.isdigit() for char in Pw))
print("Is alphanumeric:", Pw.isalnum())
print("Is all lower:", Pw.islower())
print("Is all upper:", Pw.isupper())


```
